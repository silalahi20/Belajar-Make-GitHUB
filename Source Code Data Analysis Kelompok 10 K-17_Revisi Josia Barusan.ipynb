# IMPORT (PANDAS & MATPLOTLIB) DAN MEMBACA DATA

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("indonesian_movies.csv")
df
```

# DIMENSI DATASET

```python
print("Dimensi dataset adalah: " + str(len(df)) + " baris x " + str(len(df.columns)) + " kolom.")

print("Nama-nama kolom pada dataset: ")
kolom = list(df.columns)
for i in range(len(kolom)):
    print(str(i+1) + ". " + str(kolom[i]) + " (" + str(df.dtypes[str(kolom[i])]) + ")")
```

# MENGUBAH DATA TYPE DARI BEBERAPA KOLOM AGAR BISA DIGUNAKAN SEBAGAI DATA KUANTITATIF

```python
# menghilangkan kata "min" dari dataset agar bisa di convert menjadi data kuantitatif
df["runtime"]=df["runtime"].str.replace("min","")
```

```python
# menghilangkan koma yang menandakan ribuan agar data bisa digunakan dengan benar
df["votes"]=df["votes"].str.replace(",","")
```

```python
# mengubah tipe data ke numeric agar bisa digunakan sebagai data kuantitatif
df["votes"]=pd.to_numeric(df["votes"])
df["runtime"]=pd.to_numeric(df["runtime"])
df['year']=pd.to_numeric(df['year'])
```

```python
dfnull=df.isnull().sum()
print("1. Judul Film")
print("Judul Film adalah unik untuk masing-masing film")
print("Jumlah data kosong: " + str(dfnull["title"]))
print()
print("2. Year(Tahun Rilis Film tesebut)")
print("Range: " +str(df["year"].min()) + " - " + str(df["year"].max()))
print("Jumlah data kosong: " + str(dfnull["year"]))
print()
print("3. Description(Sinopsis Film)")
print("Sinopsis adalah unik untuk masing-masing Film")
print("Jumlah data kosong: " + str(dfnull["description"]))
print()
print("4. Genre Film")
print("Genre adalah tipe kategorikal dengan kategori sebagai berikut: " + str(list(df["genre"].unique())))
print("Jumlah data kosong: " + str(dfnull["genre"]))
print()
print("5. Rating(Usia)")
print("Rating adalah tipe kategorikal dengan kategori sebagai berikut: " + str(list(df["rating"].unique())))
print("Jumlah data kosong: " + str(dfnull["rating"]))
print()
print("6. Users Rating(Penilaian dari pengguna IMdb)")
print("Range rating film: " + str(df["users_rating"].min()) + " - " + str(df["users_rating"].max()))
print("Jumlah data kosong: " + str(dfnull["users_rating"]))
print()
print("7.Votes")
print("Votes adalah data kuantitatif berisi jumlah orang yang memberikan penilaian pada tiap film")
print("Range jumlah votes: " + str(df["votes"].min()) + " - " + str(df["votes"].max()))
print("Jumlah data kosong: " + str(dfnull["votes"]))
print()
print("8. Languages(Bahasa pada Film)")
print("Bahasa adalah tipe kategorikal dengan kategori sebagai berikut: " + str(list(df["languages"].unique()))) 
print("Jumlah data kosong: " + str(dfnull["languages"]))
print()
print("9. Directors ")
print("Sutradara adalah unik untuk masing-masing film walaupun ada beberapa sutradara yang merilis beberapa film")
print("Jumlah data kosong: " + str(dfnull["directors"]))
print()
print("10. Actors")
print("Aktor dan aktris pada masing-masing film adalah unik")
print("Jumlah data kosong: " + str(dfnull["actors"]))
print()
print("11. Runtime")
print("Runtime adalah durasi masing-masing film")
print("Range durasi film dari dataset ini (dalam menit) adalah sebagai berikut: "+ str(df["runtime"].min()) + " - " + str(df["runtime"].max()))
print("Jumlah data kosong: " + str(dfnull["runtime"]))
```

# KESIMPULAN DARI KARAKTERISTIK DATA:
Ada beberapa kolom yang harus dibersihkan (Data Cleansing) yaitu genre,directors, dan runtime. Untuk data pada kolom Rating tidak kami gunakan karena data kosong mencapai 896 (+- 70%)

# DATA CLEANSING

```python
# menghapus column rating dan description karena tidak digunakan
df = df.drop('rating', axis=1)
df = df.drop('description', axis=1)
```

```python
# menghapus baris data yang memiliki elemen runtime / directors / genre yang null
print("Hapus beberapa data sesuai yang sudah disebutkan di atas")
df = df.dropna()
print("Setelah data yang kosong dihapus maka dataset df menjadi " + str(len(df)) + " baris")
df
```

```python
#tipe data
print("Nama-nama kolom pada dataset: ")
kolom = list(df.columns)
for i in range(len(kolom)):
    print(str(i+1) + ". " + str(kolom[i]) + " (" + str(df.dtypes[str(kolom[i])]) + ")")
```

## PERKEMBANGAN GENRE FILM INDONESIA 

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("indonesian_movies.csv")
df3= df.sort_values(["year","title"],ascending=[True,True])

df3.groupby("year")["genre"].value_counts()[:37].unstack().plot(kind="bar", figsize=(16,8),stacked=True, legend=True)
plt.ylabel("Jumlah Film Tiap Genre")
plt.title("Produksi film pada tahun 1926 - 1979")
plt.show()
df3.groupby("year")["genre"].value_counts()[37:90].unstack().plot(kind="bar", figsize=(16,8),stacked=True, legend=True)
plt.ylabel("Jumlah Film Tiap Genre")
plt.title("Produksi film pada tahun 1980 - 1990")
plt.show()
df3.groupby("year")["genre"].value_counts()[90:119].unstack().plot(kind="bar", figsize=(16,8),stacked=True, legend=True)
plt.ylabel("Jumlah Film Tiap Genre")
plt.title("Produksi Film pada tahun 1991 - 2000")
plt.show()
df3.groupby("year")["genre"].value_counts()[119:170].unstack().plot(kind="bar", figsize=(16,8),stacked=True, legend=True)
plt.ylabel("Jumlah Film Tiap Genre")
plt.title("Produksi Film pada tahun 2001 - 2010")
plt.show()
df3.groupby("year")["genre"].value_counts()[170:].unstack().plot(kind="bar", figsize=(16,8),stacked=True, legend=True)
plt.ylabel("Jumlah Film Tiap Genre")
plt.title("Produksi Film pada tahun 2011 - 2020")
plt.show()
```

### KESIMPULAN :
Film Indonesia genre-nya mulai bervariatif dan berkembang mulai tahun 2011 sampai 2022 walaupun dengan jumlah yang fluktuatif.
Puncaknya adalah tahun 2019, walaupun genrenya kurang bervariatif.
